Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import lambdify
```

```python
def gradient_descent(alpha, start, max_iterations):
    x = sp.symbols('x')
    objective = (x + 3)**2  
    derivative = sp.diff(objective, x)
    gradient = sp.lambdify(x, derivative)  

    x_list = []
    x_val = start
    x_list.append(x_val)

    for i in range(max_iterations):
        gradient_val = gradient(x_val)
        x_val = x_val - alpha * gradient_val
        x_list.append(x_val)

    return x_list, objective

alpha = 0.1
start = 2
max_iter = 150

x_val, objective = gradient_descent(alpha, start, max_iter)
print("Minima:", x_val[-1])

x = sp.symbols('x')
x_cor = np.linspace(-4, 3, 100)
plt.plot(x_cor, lambdify(x, objective)(x_cor), label="Objective Function")

x_arr = np.array(x_val)
plt.plot(x_arr, lambdify(x, objective)(x_arr), '.-', color='red', label="Gradient Descent Path")
plt.legend()
plt.show()
```
